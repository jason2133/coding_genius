```
def grade(x):
  if x > 90 :
    print("A")
  
  elif 80 <= x < 90:
    print("B")
  
  elif 70 <= x < 80:
    print("C")
    
  elif 60 <= x < 70:
    print("D")
  
  else:
  # if x < 60: 과 같음
    print("F")
```

```
grade(95)
```

```
grade(85)
```
