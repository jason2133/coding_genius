```
# 클래스 정의
class Fourcal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    def sum(self):
        result = self.first + self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
```

```
# 데이터 정의
a = Fourcal()
a.setdata(6, 9)
```

```
# 덧셈
print(a.sum())
```

```
# 뺄셈
print(a.sub())
```

```
# 곱셈
print(a.mul())
```

```
# 나눗셈
print(a.div())
```
