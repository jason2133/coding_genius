```
# a*x + b = 0

a, b = map(int, input().split())

x = (-b) / a

print("1차방정식 : %dx + %d = 0" % (a, b))
print(x)
```

```
def onecha(a, b):
  x = (-b) / a
  print("1차방정식 : %dx + %d = 0" % (a, b))
  print(x)
```

```
onecha(3, 3)
```

```
def onecha2(a, b, c):
    x = (c - b) / a
    print("1차방정식 : %dx + %d = %d" % (a, b, c))
    print(x)
```

```
onecha2(2, 2, 5)
```
